# Categorical plotting using Seaborn

https://seaborn.pydata.org/tutorial/categorical.html

We've seen a number of plots meant to show the characteristics of continuouos variables in exploratory data analysis,  these are few meant for examining
categorical variables

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python

sns.set_theme(style="darkgrid")
tips = sns.load_dataset("tips")
ax = sns.pointplot(x="time", y="total_bill", data=tips)
```

```python
# tips is Seaborn example data set,  it looks at tips are a restaurant
# this is back when smoking was legal in restaurants

tips.head()
```

## Question,  what is this plot showing us?  The dots are the means, the bar height looks like a range measure-it is described as an error bar

Check the min and max total bill for dinner and lunch

Also check the standard deviation for the dinner and lunch tips



```python
ax = sns.pointplot(x="time", y="total_bill", hue="smoker",data=tips)
```

```python
#add a bit of an offset
ax = sns.pointplot(x="time", y="total_bill", hue="smoker",data=tips, dodge=0.03)
```

```python
# bootstrap resamplign to estimate the confidence intervals

# this is using a bootstrap method instead of the standard estimater for error in the mean.  If the distribution was highly non-normal the bookstrap method may be more effective

ax = sns.pointplot(x="time", y="total_bill", hue="smoker",data=tips, dodge=0.03,n_boot=100)
```

```python
# more plot control

ax = sns.pointplot(x="time", y="total_bill", hue="smoker",
                   data=tips,
                   markers=["o", "x"],
                   linestyles=["-", "--"])
```

```python
ax = sns.pointplot(x="day", y="tip", data=tips, capsize=.2,hue="smoker")
```

Notice the shift here in the impact of smokers on tips.   This plot is showing both lunch and dinner,  so there is something complicated going on here
related to the number of smokers and the ratio of lunches per day to the dinners per day

```python
g = sns.catplot(x="sex", y="total_bill",
                hue="smoker", col="time",
                data=tips, kind="point",
                dodge=True,
                height=4, aspect=.7);
```

```python
g = sns.catplot(x="day", y="total_bill",
                hue="smoker", col="time",
                data=tips, kind="point",
                dodge=True,
                height=4, aspect=.7);
```

## Count Plots

```python
# using the classic titanic data set

sns.set_theme(style="darkgrid")
titanic = sns.load_dataset("titanic")
```

```python
# we have a lot of categorical variables here,   survived, passenger class, sex, embarked, class, who, deck, embarktown, alone
# making this a nice example for working with categorical information

titanic.head()
```

```python
# counts by passenger class

ax = sns.countplot(x="class", hue="who", data=titanic)
```

```python
# here's the cool one...

g = sns.catplot(x="class", hue="who", col="survived",
                data=titanic, kind="count",
                height=4, aspect=.7);
```
